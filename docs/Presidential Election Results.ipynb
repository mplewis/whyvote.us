```python
def trim_to(rows, year):
    col = rows[0].index(str(year))
    return [r[:col] for r in rows]

trim_to(raw, 1988)[:5]
```

```python
trimmed = trim_to(raw, 1988)
header = trimmed[0]
states_only = trimmed[3:]
[s[0] for s in states_only]
```

```python
def years_represented(header):
    years = []
    for cell in header:
        try:
            int(cell)
            years.append(cell)
        except ValueError:
            continue
    return years
            
years_represented(header)
```

```python
def parse_pct(s):
    return float(s.split('%')[0])

def voting_history(state_row, years):
    y = years.copy()
    history = []
    is_margin = False
    is_dem_pct = True
    is_rep_pct = False
    for col, cell in enumerate(state_row[1:]):
        if is_dem_pct:
            dem_pct = parse_pct(cell)
            is_dem_pct = False
            is_rep_pct = True
        elif is_rep_pct:
            rep_pct = parse_pct(cell)
            is_rep_pct = False
        elif is_margin:
            cook_pvi_dem = int(cell)
            if not dem_win:
                cook_pvi_dem = -cook_pvi_dem
            is_margin = False
            is_dem_pct = True
            year = y.pop(0)
            history.append({'dem_win': dem_win,
                            'cook_pvi_dem': cook_pvi_dem,
                            'dem_pct': dem_pct,
                            'rep_pct': rep_pct,
                            'dem_win_margin': round(dem_pct - rep_pct, 1),
                            'year': year})
        elif cell == 'D+':
            dem_win = True
            is_margin = True
        elif cell == 'R+':
            dem_win = False
            is_margin = True
        
    return history

from pprint import pprint

trimmed = trim_to(raw, 1988)
header = trimmed[0]
years = years_represented(header)

for state in states_only[3:7]:
    print(state[0])
    pprint(voting_history(state, years))
```

```python
states_only = trimmed[3:]
histories = []
for state in states_only:
    abbr = state[0]
    history = voting_history(state, years)
    histories.append([abbr, history])
pprint(histories[3:7])
```

```python
from statistics import mean, stdev

def mean_sd(history):
    margins = [i['dem_win_margin'] for i in history]
    return mean(margins), stdev(margins)
```

```python
import scipy.stats
def est_prob_votes_dem(mean, sd):
    # mean > 0: percentage points tend to swing democrat
    return scipy.stats.norm(-mean, sd).cdf(0)

est_prob_votes_dem(2, 5)
```

```python
def to_slug(name):
    return name.lower().replace(' ', '-')

to_slug('Washington DC')
```

```python
trimmed = trim_to(raw, 1988)
header = trimmed[0]
states_only = trimmed[3:]

years = years_represented(header)
print(years)
state_record = []
for state in states_only:
    name = state[0]
    history = voting_history(state, years)
    state_record.append({'name': name, 'history': history,
                         'slug': to_slug(name)})
pprint(state_record[3:7])
```

```python
for state in state_record[3:7]:
    m, s = mean_sd(state['history'])
    epvd = est_prob_votes_dem(m, s)
    pprint(state)
    print(epvd)
```

```python
def is_swing(single_vote):
    return abs(single_vote['dem_win_margin']) <= 3

closest = []
for r in state_record:
    close = len(list(filter(is_swing, r['history'])))
    closest.append((r['name'], close))

sorted(closest, key=lambda p: p[1], reverse=True)
```

```python
import json

# data from http://data.cnn.com/ELECTION/2016/full/P.full.json
with open('/Users/mplewis/Downloads/2016_election.json') as f:
    raw_2016 = json.load(f)
    
raw_2016.keys()
```

```python
def record_for(state_name):
    for state in state_record:
        if state['name'] == state_name:
            return state
        if state['name'] == 'Washington DC' and state_name == 'Dist. of Columbia':
            return state

record_for('Dist. of Columbia')
```

```python
for race in raw_2016['races']:
    state = race['state']
    target = record_for(state)
    if target is None:
        continue
    dem_pct = None
    rep_pct = None
    dem_win = None
    for cand in race['candidates']:
        party = cand['party']
        if party == 'D':
            dem_pct = float(cand['pctDecimal'])
            if cand['winner']:
                dem_win = True
        elif party == 'R':
            rep_pct = float(cand['pctDecimal'])
            if cand['winner']:
                dem_win = False
    record_2016 = {
        'year': '2016',
        'dem_win': dem_win,
        'dem_pct': dem_pct,
        'rep_pct': rep_pct,
        'dem_win_margin': round(dem_pct - rep_pct, 1)
    }
    print(state)
    pprint(record_2016)
    target['history'].insert(0, record_2016)
```
